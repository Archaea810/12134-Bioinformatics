```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('WHO-COVID-19-global-data.csv')
df.loc[df['New_deaths'] < 0, 'New_deaths'] = 0
df.loc[df['New_cases'] < 0, 'New_cases'] = 0
```

```python
df[df['Country_code']=='FR']
```

```python
df[df['Date_reported']=='2020-03-11'][df['Country_code']=='FR'] #Original pandemic declaration
```

```python
df[df['Date_reported']=='2021-01-11'][df['Country_code']=='FR'] #Gamma
```

```python
df[df['Date_reported']=='2021-04-04'][df['Country_code']=='FR'] #Delta
```

```python
df[df['Date_reported']=='2021-11-26'][df['Country_code']=='FR'] #Omicron
```

```python
(df[df['Country_code']=='FR']['Cumulative_deaths'] / df[df['Country_code']=='FR']['Cumulative_cases']).dropna()
```

```python
plt.plot((df[df['Country_code']=='FR']['New_cases']).dropna())
plt.title('New cases of SARS-CoV-2 in FRANCE')
#Pandemic Declaration
plt.axvline(x=86180, color='k', linestyle='--')
#Alpha
plt.axvline(x=86462, color='k', linestyle='--')
#Gamma
plt.axvline(x=86486, color='k', linestyle='--')
#Delta
plt.axvline(x=86569, color='k', linestyle='--')
#Omicron
plt.axvline(x=86805, color='k', linestyle='--')
plt.savefig('./Figures/FR_nc.png', dpi=500)
plt.show()
```

```python
plt.plot(df[df['Country_code']=='FR']['New_deaths'])
plt.title('New deaths of SARS-CoV-2 in FRANCE')
#Pandemic Declaration
plt.axvline(x=86180, color='k', linestyle='--')
#Alpha
plt.axvline(x=86462, color='k', linestyle='--')
#Gamma
plt.axvline(x=86486, color='k', linestyle='--')
#Delta
plt.axvline(x=86569, color='k', linestyle='--')
#Omicron
plt.axvline(x=86805, color='k', linestyle='--')
plt.savefig('./Figures/FR_nd.png', dpi=500)
plt.show()
```

```python
plt.plot((df[df['Country_code']=='FR']['New_deaths'] / df[df['Country_code']=='FR']['New_cases']).dropna())
plt.title('Fatality rate of SARS-CoV-2 in FRANCE')
#Pandemic Declaration
plt.axvline(x=86180, color='k', linestyle='--')
#Alpha
plt.axvline(x=86462, color='k', linestyle='--')
#Gamma
plt.axvline(x=86486, color='k', linestyle='--')
#Delta
plt.axvline(x=86569, color='k', linestyle='--')
#Omicron
plt.axvline(x=86805, color='k', linestyle='--')
plt.savefig('./Figures/FR_dr.png', dpi=500)
plt.show()
```
