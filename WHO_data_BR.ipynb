```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('WHO-COVID-19-global-data.csv')
df.loc[df['New_deaths'] < 0, 'New_deaths'] = 0
df.loc[df['New_cases'] < 0, 'New_cases'] = 0
```

```python
plt.plot((df[df['Country_code']=='BR']['New_cases']).dropna())
plt.title('New cases of SARS-CoV-2 in BRAZIL')
#Pandemic Declaration
plt.axvline(x=33556, color='k', linestyle='--')
#Alpha
plt.axvline(x=33838, color='k', linestyle='--')
#Gamma
plt.axvline(x=33862, color='k', linestyle='--')
#Delta
plt.axvline(x=33945, color='k', linestyle='--')
#Omicron
plt.axvline(x=34181, color='k', linestyle='--')
plt.savefig('./Figures/BR_nc.png', dpi=500)
plt.show()
```

```python
plt.plot(df[df['Country_code']=='BR']['New_deaths'])
plt.title('New deaths of SARS-CoV-2 in BRAZIL')
#Pandemic Declaration
plt.axvline(x=33556, color='k', linestyle='--')
#Alpha
plt.axvline(x=33838, color='k', linestyle='--')
#Gamma
plt.axvline(x=33862, color='k', linestyle='--')
#Delta
plt.axvline(x=33945, color='k', linestyle='--')
#Omicron
plt.axvline(x=34181, color='k', linestyle='--')
plt.savefig('./Figures/BR_nd.png', dpi=500)
plt.show()
```

```python
plt.plot((df[df['Country_code']=='BR']['New_deaths'] / df[df['Country_code']=='BR']['New_cases']).dropna())
plt.title('Fatality rate of SARS-CoV-2 in BRAZIL')
#Pandemic Declaration
plt.axvline(x=33556, color='k', linestyle='--')
#Alpha
plt.axvline(x=33838, color='k', linestyle='--')
#Gamma
plt.axvline(x=33862, color='k', linestyle='--')
#Delta
plt.axvline(x=33945, color='k', linestyle='--')
#Omicron
plt.axvline(x=34181, color='k', linestyle='--')
plt.savefig('./Figures/BR_dr.png', dpi=500)
plt.show()
```
